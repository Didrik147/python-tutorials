```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.tag import pos_tag
import json

text = open('HarryPotter.txt')
text = text.read()
```

```python
# Create a frequency dictionary for all the words in Harry Potter
with open('words.json', 'r') as fp:
    words =  json.load(fp)

# Use Frequency Distribution function from NLTK
freq_distribution = FreqDist(words)
freq_d = dict(freq_distribution)
print(freq_d)
```

```python
# Use sorted() to create a ranked list of words
# Print out each word and the number of times it repeats
ranked_list = sorted(freq_d, key=freq_d.get, reverse = True)
for i in range(0,len(freq_d)):
    key = ranked_list[i]
    value = freq_d[key]
    print(key,value)
```

```python
# Print ranking after eliminating stop words
stop_words = set(stopwords.words("english"))
words = set(words)
unique_words = []

for word in ranked_list:
    if word.lower() not in stop_words:
        print(word,':',freq_d[word])

ranked_list.remove('Well')
ranked_list.remove('Oh')
        
```

```python
# Print the ranking of all proper nouns
tagged_text = pos_tag(text.split())
proper_nouns = [word for word,pos in tagged_text if pos == 'NNP']

filtered_d = {}
# Remove any anomalies

for word in ranked_list:
    if word in proper_nouns and word.istitle() and word.lower() not in stop_words:
        filtered_d[word] = freq_d[word]
        print(word,':',freq_d[word])
```

```python
# Create and rank a list of verbs in the book
verbs = [word for word,pos in tagged_text if pos == 'VB']
# Eliminate the words that are also classified other than verbs
not_verbs = [word for word,pos in tagged_text if pos !='VB']

for v in verbs:
    if v in not_verbs:
        verbs.remove(v)

verb_d = {}


for word in ranked_list:
    if word in verbs:
        verb_d[word] = freq_d[word]
        print(word,':',freq_d[word])
```

```python
%matplotlib inline
# Put the dictionary into a Data Frame and print out a bar graph
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame.from_dict(filtered_d, orient = 'index').iloc[0:30]

print(df)

df.plot.bar(figsize = (20,10), title= 'HARRY POTTER PROTAGONISTS')
#df.plot(figsize = (10,10)).bar()
```
