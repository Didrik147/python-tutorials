```python
# Import libraries
import json
import nltk
from nltk.probability import FreqDist
from nameparser.parser import HumanName
from nltk.corpus import wordnet

text = open('HarryPotter.txt')
text = text.read()
```

```python
#Analyses text and extracts two word human names
'''function source: https://stackoverflow.com/questions/20290870/improving-the-extraction-of-human-names-with-nltk'''
def get_human_names(text):
    person_list = []
    person_names=person_list
    tokens = nltk.tokenize.word_tokenize(text)
    pos = nltk.pos_tag(tokens)
    sentt = nltk.ne_chunk(pos, binary = False)

    person = []
    name = ""
    for subtree in sentt.subtrees(filter=lambda t: t.label() == 'PERSON'):
        for leaf in subtree.leaves():
            person.append(leaf[0])
        if len(person) > 1: #avoid grabbing lone surnames
            for part in person:
                name += part + ' '
            if name[:-1] not in person_list:
                person_list.append(name[:-1])
            name = ''
        person = []
    return person_list
```

```python
human_names = get_human_names(text)
print(human_names)
```

```python
# Print out the ranking of full names
from nltk.probability import FreqDist
with open('words.json', 'r') as fp:
    words =  json.load(fp)

#Use Frequency Distribution function from NLTK
realname_d = {}
for name in human_names:
    realname_d[name] = text.count(name)


ranked_list = sorted(realname_d, key=realname_d.get, reverse = True)

for word in ranked_list:
    print(word,':',realname_d[word])
```
