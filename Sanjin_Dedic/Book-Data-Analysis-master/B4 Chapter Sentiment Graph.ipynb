```python
from textblob import TextBlob
import json

text = open('HarryPotter.txt')
text = text.read()
```

```python
# Load chapters file
with open('chapters.json', 'r') as fp:
    chapters =  json.load(fp)
    
# Create a dictionary where chapter num is key and chapter is the value
chapter_sentiment = {}
chapter = 0 #change this to make chapters align

for c in chapters:
    chapter_sentiment[str(chapter)] = TextBlob(c).polarity
    chapter+=1
```

```python
print(chapter_sentiment)
```

```python
#Print out the bar graph of chapter to chapter progress

import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame.from_dict(chapter_sentiment, orient = 'index').iloc[1:18]
#print(df)

df.plot.bar(figsize = (20,10))
#df.plot(figsize = (10,10)).bar()
```

```python
# Print out the line graph for the same data
df.plot.line(figsize = (20,10))
```
